```python
import matplotlib.pyplot as plt
import numpy as np
import random
np.random.randint(10,50,(10))
x=[16, 46, 45, 17, 24, 26, 26, 33, 18, 46]
color=["black"]
label=["fisrt"]
plt.hist(x,color=color,edgecolor="r",range=(0,200),cumulative=-1,bottom=12,align="mid",orientation="vertical",histtype="bar",rwidth=0.4,log=True,label=label)
#types of histogram...step,step fill, barstacked,bar)
#rwidth
#error show(iin third use)
#grp(draw a line)
plt.xlabel("X-axis",fontsize=10)
plt.ylabel("Y-axis",fontsize=10)      
plt.title("Data Visalization(Histogram)")
plt.grid()
plt.axvline("25",color="green",label="line")
plt.legend()
plt.show()
```
