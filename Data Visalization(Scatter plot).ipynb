```python
#pip install matplot library(Nmerical Nmerical)
import  matplotlib.pyplot as plt
x=[1,2,3]
y=[2,3,3]
color=["purple","green","yellow"]
size=[100,200,300]
alpha=[0.1,0.3,0.6]
label=["first"]
#color & marker show as a star,edgecolor can change
plt.scatter(x,y,color=color,marker="*",edgecolor="y",s=size,alpha=alpha,label="first")
#Tile(show x & y)
plt.legend("Scatter")
plt.title("Scatter Plot")
#font size=any vale
plt.xlabel("x-axis",fontsize="9")
plt.ylabel("y-axis",fontsize="9")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[3,4,5,6]
color=["yellow","red","black","pink"]
size=[122,123,124,126]
label=["sd"]
alpha=[0.1,0.2,0.3,0.4]
plt.scatter(x,y,c=color,s=size,label="label",alpha=alpha)
plt.legend("Scatter Plot")
plt.title("Scatter Plot")
plt.xlabel("x-axis",fontsize=10)
plt.ylabel("y-axis",fontsize=10)
plt.show()
```
