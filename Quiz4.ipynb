```python
#QNo1)What is machine learning?
Machine Learning is a field of compter science that uses  statistical techniques to give 
computer system the ability to learn with data without being expilicity program.
```

```python
#QNO2)What are the three main types of machine learning?
The three main types of machine learning are:
(i)Supervised Learning: The model is trained on labeled data (input-output pairs) to make predictions.
(ii)Unsupervised Learning: The model works with unlabeled data and tries to find patterns or groupings (e.g., clustering).
(iii)Reinforcement Learning: The model learns by interacting with an environment and receiving feedback through rewards or penalties.
```

```python
#QN03)Define supervised learning and provide an example.
Supervised learning is a type of machine learning where a model is trained on labeled data to predict outcomes based on input.
Example:Predicting the price of a house based on its size
```

```python
#QN04)Define unsupervised learning and provide an example.
Unsupervised learning is a type of machine learning where the model is given data without any labels
to find patterns or groupings in the data.
Example:grouping similar fruits based on their characteristics, like size, color, and shape, without knowing the fruit names.
```

```python
#QN05)What is reinforcement learning?
Reinforcement learning is when an agent learns to make decisions by receiving rewards or penalties.
Example:teaching a dog to sit by rewarding it with treats.
```

```python
#QN06)What is the difference between classification and regression?
Problems of Supervised learning.
Classification is when the output is a category or label (e.g: predicting whether an email is "spam" or "not spam").
Regression is when the output is a continuous value or number (e.g: predicting the price of a house based on its size).
```

```python
#QN07)What is overfitting in machine learning?
Overfitting is when a model learns the details of the training data too well, including the noise, making it perform poorly on new data.
Its like memorizing answers instead of learning the material.
```

```python
#QN08)What is underfitting in machine learning?
Underfitting in machine learning happens when a model is too simple to capture the patterns in the data, leading to poor performance on both the training data and new data.
Its like trying to solve a complex problem with too basic a tool.
```

```python
#QN09)What is a feature in machine learning?
A feature in machine learning is an individual piece of data used by the model to make predictions.
Example: the size of a house in price prediction.
```

```python
#QN010)What is a label in supervised learning?
A label in supervised learning is the correct output or target value that the model is trying to predict based on the input features.
For example: in a spam email classifier, the label would be "spam" or "not spam."
```

```python
#QN011)Name one common algorithm used for classification tasks.
Decision Tree.
```

```python
#QN012)Name one common algorithm used for regression tasks.
Linear Regression
```

```python
#QN013)What is the purpose of the k-means clustering algorithm?
The k-means clustering algorithm groups data into k clusters based on similarity
minimizing the distance between data points within each cluster.
```

```python
#QNO8)What is the difference between df.loc[] and df.iloc[]?
df.loc[]: Uses label-based indexing (row labels and column names).
df.iloc[]: Uses integer-based indexing (row and column positions).
```

```python
#QNO8)....Code....
import pandas as pd
# Sample DataFrame
data = {"Name": ["Ali","Eisha","Romana"], 
        "Age": [25, 30, 35]}
df = pd.DataFrame(data, index=["a", "b", "c"])

# Using df.loc[] (label-based)
print(df.loc["b"])  # Selects the row with index label "b"
# Using df.iloc[] (position-based)
print(df.iloc[1])  # Selects the row at position 1 (second row)
```

```python
#QN010)How do you group data in a DataFrame using groupby()?
import pandas as pd

# Sample DataFrame
data = {"Department": ["HR", "IT", "HR", "IT", "Finance"],
        "Employee": ["Ali", "Bilal", "Charmy", "David", "Eva"],
        "Salary": [50000, 60000, 55000, 70000, 65000]}
df = pd.DataFrame(data)
# Grouping by "Department" and calculating the average salary
grouped_df = df.groupby("Department")["Salary"].mean()
print(grouped_df)
```

```python
#QN05)What is the difference between sns.histplot() and sns.kdeplot()?
sns.histplot() creates a histogram to show the distribution of data while sns.kdeplot() creates a Kernel Density 
Estimate plot to show the smooth probability density function of the data.
sns.histplot() will show the histogram (bar plot).
sns.kdeplot() will show the KDE (smooth line plot).
```

```python
#QN05)....Code....
import seaborn as sns
import matplotlib.pyplot as plt
# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5, 6]
# Creating a histogram
sns.histplot(data, bins=6, kde=False, color="blue", label="Histogram")
# Creating a KDE plot
sns.kdeplot(data, color="red", label="KDE")
# Adding labels and title
plt.title("Difference Between Histogram and KDE")
plt.xlabel("Value")
plt.ylabel("Density")
# Display the plot
plt.legend()
plt.show()
```
